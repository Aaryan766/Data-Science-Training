```python
from googlesearch import search

title = 'AI-ML Complete Tutorial PDF'
data = search( title, num_results = 50)

all_data = []

for i in data:
    all_data.append(i + '\n')

file = open(f'{title}.txt', mode = 'w')
file.writelines(all_data)
file.close()
print('Done')
```

```python
from googlesearch import search

title = 'AI-ML Complete Tutorial PDF'
data = search( title, num_results = 50)

all_data = []

for i in data:
    all_data.append(i + '\n')

# Mode = x : it only creates new file(FileExistsError) 

file = open(f'{title}.txt', mode = 'x')
file.writelines(all_data)
file.close()
print('Done')
```

```python
url = '''https://picsum.photos/1500/1080'''
import requests as r 
n = 10
for i in range(n):
    image_data = r.get(url).content
    f = open(f'image_{i}.jpg', mode = 'wb')
    f.write(image_data)
    f.close()
    from IPython import display as d
    display(d.Image(f'image_{i}.jpg', width = 200, height = 200))
```
