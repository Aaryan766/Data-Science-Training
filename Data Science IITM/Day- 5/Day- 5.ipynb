```python
def show_records(db,table):
    
    query = f'''select * from {db}.{table};'''
    
    cur.execute(query)
        
    records = cur.fetchall()
    
    all_columns=[]
    for i in cur.description:
        all_columns.append(i[0])
        
    
    import pandas as pd
    df = pd.DataFrame(records,columns = all_columns)
    return df
```
